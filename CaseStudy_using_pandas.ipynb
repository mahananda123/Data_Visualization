###Data Visualization
It is hard to digest many data at the same time. Data visualizaiton is a way to make the process simple and straightforward.

Because Data Visualization is so important and essential in today's business, pandas as a datadriven package provides built-in plotting functions. We will learn some basic usage of them today.

Data visualization techniques are
1. Scatter plots
2. line plots
3. Area plots
4. Bar plots
5. Histograms
6. Box plot
7. Pie Charts
8. color map etc..


```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/ds-salaries.csv')
df.head(15)
```

```python
df.info()
```

```python
df.describe()
```

```python
#scatter plot
df.plot(x='remote_ratio',y='salary_in_usd',kind='scatter')
```

```python
#line plot
df.plot(y='remote_ratio',kind='line')
```

```python
df.plot(x='salary',y='remote_ratio',kind='line')
```

```python
#Area plot
df.plot(x='salary',y='remote_ratio',kind='area')
```

```python
df.plot(x='salary_in_usd',y='remote_ratio',kind='area')
```

```python
#bar plots
df.plot(x='salary_in_usd',y='remote_ratio',kind='bar')
```

```python
#bar plots
df.plot(x='salary_in_usd',y='remote_ratio',kind='barh')
```

```python
#Histograms
df['salary'].plot(kind='hist')
```

```python
df['salary'].plot(kind='hist', bins=1000)
```

```python
#box plot
df['salary_in_usd'].plot(kind='box')
```

```python
df['salary_in_usd'].plot(kind='kde')
```

```python
df['salary_in_usd'].value_counts().plot(kind='pie')
```

```python
df.plot.scatter(x = 'salary', y = 'salary_in_usd', c = 'remote_ratio')
```
