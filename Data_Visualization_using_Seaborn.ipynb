```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips
```

###Relational Plots

```python
sns.relplot(data=tips,x='total_bill',y='tip')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',col='day')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',col='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='size',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='size',size='size',sizes=(15,200))
```

```python
sns.relplot(data=tips,x='day',y='total_bill')
```

```python
sns.relplot(data=tips,x='day',y='total_bill',kind='line')
```

```python
sns.relplot(data=tips,x='day',y='total_bill',kind='line',hue='sex')
```

```python
sns.relplot(data=tips,x='day',y='total_bill',kind='line',hue='sex',col='time')
```

###Distribution Plots

```python
sns.displot(data=tips,x='total_bill').set(title='distribution of total bill',xlabel='Total Bill')
```

```python
sns.displot(data=tips,x='total_bill',col='sex').set(title='distribution of total bill')
```

```python
sns.displot(data=tips,x='total_bill',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde')
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',multiple='stack')
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',multiple='stack',col='day')
```

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde')
```

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde',hue='sex',rug=True)
```

```python
sns.displot(data=tips,x='total_bill',kind='ecdf')
```

```python
sns.displot(data=tips,x='total_bill',kind='ecdf',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='ecdf',hue='sex',col='day')
```

###Categorical Plots

```python
sns.catplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True,inner='quartile')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='bar')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='bar',hue='sex')
```

###Joint plots

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='kde')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='scatter')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='resid')
```
