###Documentation:  https://matplotlib.org

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Economy_of_US.csv')
df.head(15)
```

###Simple plot


```python
plt.plot(df['Year'],df['GDP_Growth'])
plt.xlabel('YEAR')
plt.ylabel("GDP GROWTH")
plt.title('Economy of US')
```

###Change markers

```python
plt.plot(df['Year'],df['GDP_Growth'],'^')
plt.xlabel('YEAR')
plt.ylabel("GDP GROWTH")
plt.title('Economy of US')
```

```python
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v']
for m in markers:
    print(m)
    plt.plot(df['Year'],df['GDP_Growth'],m)
    plt.xlabel('YEAR')
    plt.ylabel("GDP GROWTH")
    plt.title('Economy of US')
    plt.show()
```

###Change line style and color

```python
line=['-',':','--','-.']
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v']
color = ['r','g','b','c','m','y','k','w']
for l,m,c in zip(line,markers,color):
    print(l,m,c)
    x=m+l+c
    plt.plot(df['Year'],df['GDP_Growth'],x)
    plt.xlabel('YEAR')
    plt.ylabel("GDP GROWTH")
    plt.title('Economy of US')
    plt.show()
```

###Change in marker size

```python
plt.plot(df['Year'], df['GDP_Growth'], marker = '^', ms = 15)
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

###Scatter plot

```python
plt.scatter(df['Year'],df['GDP_Growth'],s=df['Unemployment']*1000)
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('Economy of US')
```

```python
plt.scatter(df['Year'],df['GDP_Growth'],s=df['Unemployment']*1000,cmap='hot',alpha=0.5,c=df['Inflation'])
plt.colorbar()
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('Economy of US')
```

Bar plot

```python
plt.bar(df['Year'], df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

###Histogram

```python
plt.hist( df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

```python
plt.hist( df['GDP_Growth'],bins=30)
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

###Pie plot


```python
plt.pie(df['Inflation_Change'].value_counts(),labels=['Decrease','Increase','No change'])
plt.legend()
plt.title('Economy of US')
```

```python
plt.pie(df['Inflation_Change'].value_counts(),labels=['Decrease','Increase','No change'],explode=[0.2,0.2,0.2])
plt.legend()
plt.title('Economy of US')
```

###Box Plot

```python
plt.boxplot(df['Inflation'])
plt.ylabel('Inflation')
plt.title('Economy of US')
```

###Violin plot

```python
plt.violinplot(df['Inflation'])
plt.ylabel('Inflation')
plt.title('Economy of US')
```

###Multi plots

```python
plt.plot(df['Year'],df['GDP_Growth'],label='GDP_Growth')
plt.plot(df['Year'],df['Inflation'],label='Inflation')
plt.plot(df['Year'],df['Unemployment'],label='Unemployement')
plt.xlabel("Year")
plt.ylabel("Economy")
plt.grid()
plt.legend()
plt.title('Economy of US')
```

###chanage arrangement


```python
plt.subplot(3,1,1)
plt.plot(df['Year'],df['GDP_Growth'],label='GDP_Growth')
plt.ylabel("Economy")
plt.grid()
plt.legend()
plt.subplot(3,1,2)
plt.plot(df['Year'],df['Inflation'],label='Inflation')
plt.xlabel("Year")
plt.ylabel("Economy")
plt.grid()
plt.legend()
plt.subplot(3,1,3)
plt.plot(df['Year'],df['Unemployment'],label='Unemployement')
plt.xlabel("Year")
plt.ylabel("Economy")
plt.grid()
plt.legend()
```

```python
plt.subplot(1,3,1)
plt.plot(df['Year'],df['GDP_Growth'],label='GDP_Growth')
plt.ylabel("Economy")
plt.grid()
plt.legend()
plt.subplot(1,3,2)
plt.plot(df['Year'],df['Inflation'],label='Inflation')
plt.xlabel("Year")
plt.ylabel("Economy")
plt.grid()
plt.legend()
plt.subplot(1,3,3)
plt.plot(df['Year'],df['Unemployment'],label='Unemployement')
plt.xlabel("Year")
plt.ylabel("Economy")
plt.grid()
plt.legend()
```
