###Case study

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/ds-salaries.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

###Simple plot

```python
plt.plot(df['salary'],df['remote_ratio'],'o:m')#o->marker, : -> line style, m ->color
plt.xlabel('Salary')
plt.ylabel('Remote Ratio')
plt.title('Salary vs Remote Ratio')
plt.show()
```

###Scatter plot

```python
plt.scatter(df['salary'],df['remote_ratio'],c=df['salary_in_usd'],cmap='hot',alpha=0.5)
plt.colorbar()
plt.xlabel('Salary')
plt.ylabel('Remote Ratio')
plt.title('Salary vs Remote Ratio')
plt.show()
```

###Bar plot

```python
plt.bar(df['remote_ratio'],df['salary_in_usd'])
plt.ylabel('Salary')
plt.xlabel('Remote Ratio')
plt.title('Salary vs Remote Ratio')
plt.show()
```

###Histogram

```python
plt.hist(df['salary_in_usd'])
plt.xlabel('Salary in usd')
plt.ylabel('Frequency')
plt.title('Frequency of Salary in USD')
plt.show()
```

###Pie plot

```python
plt.pie(df['remote_ratio'].value_counts(),labels=['0','50','100'],explode=[0.1,0.1,0.1])
plt.legend()
plt.title("Remote ratio's")
plt.show()
```

###Box plot

```python
plt.boxplot(df['remote_ratio'])
plt.ylabel('remote_ratio')
plt.show()
```

###Violin plot

```python
plt.violinplot(df['remote_ratio'])
plt.ylabel('Remote Ratio')
plt.show()
```
