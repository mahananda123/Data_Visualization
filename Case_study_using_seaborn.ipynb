```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=pd.read_csv('/content/ds-salaries.csv')
tips.head()
```

```python
tips
```

###Relational plots

```python
sns.relplot(data=tips,y='salary',x='experience_level',hue='remote_ratio',col='company_size',kind='line')
```

```python
sns.relplot(data=tips,x='salary_in_usd',y='job_title',hue='remote_ratio',size='remote_ratio',sizes=(20,200),alpha=0.5,col='company_size',kind='scatter')
```

###Distribution Plots

```python
sns.displot(data=tips,y='company_location',col='company_size',hue='remote_ratio')
```

```python
sns.displot(data=tips,x='salary_in_usd',col='company_size',hue='remote_ratio',kind='kde',multiple='stack')
```

```python
sns.displot(data=tips,x='salary_in_usd',y='salary',col='company_size',hue='remote_ratio',kind='kde',multiple='stack',rug=True)
```

```python
sns.displot(data=tips,x='salary_in_usd',col='company_size',hue='remote_ratio',kind='ecdf')
```

Categorical Plots

```python

sns.catplot(data=tips,x='work_year',y='salary_in_usd',hue='remote_ratio')
```

```python

sns.catplot(data=tips,x='work_year',y='salary_in_usd',kind='box',hue='remote_ratio')
```

```python

sns.catplot(data=tips,x='work_year',y='salary_in_usd',kind='violin',hue='remote_ratio',inner='quartile')
```

```python

sns.catplot(data=tips,x='work_year',y='salary_in_usd',kind='bar',hue='remote_ratio')
```

###Joint Plots

```python
sns.jointplot(data=tips,x='salary_in_usd',y='salary',hue='remote_ratio')
```

```python
cat=['hist','hex','reg','scatter','resid']
for c in cat:
  sns.jointplot(data=tips,x='salary_in_usd',y='salary',kind=c)
```

###Pair plots

```python
sns.pairplot(data=tips)
```
